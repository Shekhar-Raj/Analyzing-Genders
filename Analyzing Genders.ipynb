# Gender Ratio in pie chart form.

# 1. Displays The total amount spent by Gender.

```python
df.groupby("Gender").sum()["Purchase"].plot(kind="pie",autopct="%.2f")
```

```python
df.groupby("Gender").mean()["Purchase"].plot(kind="pie",autopct="%.2f")
```
